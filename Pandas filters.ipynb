```python
import pandas as pd
students = pd.read_csv('D:\Костя\Downloads\StudentsPerformance.csv') #читаем табличку
students
```

```python
#Pandas series с Gender
students['gender']
```

```python
#pandas series с результатом сравнения
students['gender'] == 'female'
```

```python
students.loc[students['gender'] == 'female']
```

```python
mean_val = students['writing score'].mean() #среднее значение в Pandas Series
```

```python
#Выборка девушек у которых значение балла за письмо выше среднего
#При работе с Series необходимо использовать побитовые операторы
#и нужно указывать приоритет операций, взяв условия в скобки
students.loc[(students['writing score'] >= mean_val) & (students['gender'] == 'female')]
```

```python
#фильтрация с помощью метода query
students.query("gender == 'female' & lunch == 'standard'")

#перед названием переменной в условии query необходимо ставить @
```

```python
#выборка всех столбиков, в названиях (лейблах) которых содержится слово 'score'
#axis = 1 - поиск в названиях стобиков, axis = 0 - поиск в названиях строк (индексах)
students.filter(like = 'score', axis = 1)
```
