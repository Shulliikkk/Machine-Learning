```python
import pandas as pd
students = pd.read_csv('D:\Костя\Downloads\StudentsPerformance.csv') #читаем табличку
dota_hero = pd.read_csv('D:\Костя\Downloads\dota_hero_stats.csv')
dota_hero
```

```python
students.aggregate({'math score': 'mean', 'reading score': 'mean'})
```

```python
#найдем средние значения ключевых переменных в зависимости от пола
#сгруппируем датасет по группирующей переменной - полу
#в каждой группе найдем среднее значание
students.groupby('gender').mean()
```

```python
students.groupby('gender').aggregate({'math score': 'mean', 'reading score': 'mean'})
```

```python
students.groupby('gender', as_index = False).aggregate({'math score': 'mean', 'reading score': 'mean'})
```

```python
students.groupby(['gender', 'race/ethnicity'], as_index = False).aggregate({'math score': 'mean'})
```

```python
#найдем топ 5 математиков девушек и топ 5 математиков юношей
#отсортируем по результатам теста по математике и по полу
#ascending - по возрастанию
#сгрупируем по полу
#к каждой группе приеним head()

students.sort_values(['gender', 'math score'], ascending = False).groupby('gender').head()
```
