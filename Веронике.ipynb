```python
from random import randrange
N = 8 
Arr = [[0]*N for i in range(N)] [[0, 0, 0], [0, 5, 0], [0, 0, 0]] Arr[1][1]

for i in range(10):
    print(randrange(1, 10))
```

```python
N = 3
Arr = [[0]*N for i in range(N)]
for i in range(N):
    for j in range(N):
        Arr[i][j] = randrange(0, 10)
print(Arr)

ans = 1
max_sum = 0
max_sum_i = 0
for i in range(N):
    summ = 0
    for j in range(N):
        summ += Arr[j][i]
    if summ > max_sum:
        max_sum = summ
        max_sum_i = i
        
for i in range(N):
    ans *= Arr[i][max_sum_i]
print(ans)
```
